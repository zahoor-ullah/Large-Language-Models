```python
import pandas as pd
import numpy as np
import altair as alt
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
```

```python
import cohere
co = cohere.ClientV2("j3dnErL21HLrrwoaSskJsQpmAn2SxRc1JzJWg7yL")
```

```python
# Load the dataset to a dataframe
df_orig = pd.read_csv('https://raw.githubusercontent.com/cohere-ai/notebooks/main/notebooks/data/atis_intents_train.csv', names=['intent','query'])

# Take a small sample for illustration purposes
sample_classes = ['atis_airfare', 'atis_airline', 'atis_ground_service']
df = df_orig.sample(frac=0.1, random_state=30)
df = df[df.intent.isin(sample_classes)]
df_orig = df_orig.drop(df.index)
df.reset_index(drop=True,inplace=True)

# Remove unnecessary column 
intents = df['intent'] #save for a later need
df.drop(columns=['intent'], inplace=True)
df[:3]
```

```python
# Step 2: Turn text into Embeddings
def get_embeddings(texts, model="embed-v4.0", input_type="search_document"):
    output = co.embed(
        texts = texts,
        model = model,
        input_type = input_type,
        embedding_types = ["float"]                       
    )
    return output.embeddings.float
df["query_embeds"] = get_embeddings(df['query'].tolist())
```

```python
# Step 3: Visualize Embeddings with a Heatmap
def get_pc(arr,n):
    pca = PCA(n_components=n)
    embeds_transform = pca.fit_transform(arr)
    return embeds_transform

# Reduce embeddging to 10 principal compoonents toadi visuilzation
embeds = np.array(df['query_embeds'].tolist())
embeds_pc = get_pc(embeds,10)
embeds_pc
```

```python
# Simulating 9 embeddings, each of 768 dimensions (like OpenAI or Cohere outputs)
np.random.seed(0)
df = pd.DataFrame({
    'query_text': [
        "How to get from Logan airport to downtown Boston?",
        "Is there a shuttle from Logan airport to the city?",
        "What's the best public transport in Boston?",
        "Flight status of AA123?",
        "What is the baggage allowance for Delta?",
        "Hotels near Fenway Park?",
        "Weather in Boston this weekend?",
        "Best seafood restaurant in Boston?",
        "Tourist attractions in Boston?"
    ],
    'query_embeds': [np.random.rand(768) for _ in range(9)]
})
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(embeds_pc, annot=False, cmap='viridis', xticklabels=[f'PC{i+1}' for i in range(10)],
            yticklabels=df['query_text'].str[:50], cbar_kws={'label': 'Embedding Value'})
plt.title("Heatmap of 10-Dimensional Embeddings (PCA Compressed)")
plt.xlabel("Principal Components")
plt.ylabel("Query Snippets")
plt.tight_layout()
plt.show()
```

```python
# Reduce Embeddings to 2D with PCA
def reduce_to_2d(arr):
    pca = PCA(n_components=2)
    return pca.fit_transform(arr)
embeds_2d = reduce_to_2d(np.array(df['query_embeds'].tolist()))
embeds_2d
```

```python
df['x'] = embeds_2d[:, 0]
df['y'] = embeds_2d[:, 1]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['x'], df['y'], color='skyblue')

for i, text in enumerate(df['query_text']):
    plt.text(df['x'][i] + 0.01, df['y'][i], text[:30] + '...', fontsize=9)

plt.title("2D Visualization of Text Embeddings (PCA)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Option B: Using Altair (interactive)
import altair as alt
alt.data_transformers.disable_max_rows()

chart = alt.Chart(df).mark_circle(size=100).encode(
    x='x',
    y='y',
    tooltip='query_text'
).interactive().properties(
    width=600,
    height=400,
    title="2D Embedding Visualization (PCA)"
)

chart
```
